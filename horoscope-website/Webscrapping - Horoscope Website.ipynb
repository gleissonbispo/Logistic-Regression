# Web Scrapping from a horoscope website

```python
def horoscopo_do_dia(data):
    
    from bs4 import BeautifulSoup
    from requests import get
    from selenium import webdriver
    from time import sleep
    
    dia = data

    path = 'chromedriver.exe' #ChromeDriver 75.0.3770.90 version (http://chromedriver.chromium.org/downloads)
    options = webdriver.ChromeOptions()
    options.add_argument("--start-maximized")
    
    driver = webdriver.Chrome(path, chrome_options=options)

    signos = [
        'aries',
        'touro',
        'gemeos',
        'cancer',
        'leao',
        'virgem',
        'libra',
        'escorpiao',
        'sagitario',
        'capricornio',
        'aquario',
        'peixes'
    ]
    
    for signo in signos:
        
        url = 'https://meuhoroscopo.net/signo-' + signo + '-' + dia + '/'
        driver.get(url)
        sleep(10)
        page_source = driver.page_source
        parsed_content = BeautifulSoup(page_source, 'html.parser')


        print(f'==================================== Horóscopo {signo} - {dia} ====================================')
        print('Horóscopo Hoje: \n')
        print(parsed_content.find_all('p')[2].text.strip(), '\n')
        print('-----------------------------------------------------------------------------------------------------')
        print('Horóscopo no Amor: \n')
        print('Se você tem alguém: \n')
        print(parsed_content.find_all('p')[3].text.strip(), '\n')
        print('Se você é solteiro (a): \n')
        print(parsed_content.find_all('p')[4].text.strip(), '\n')
        print('-----------------------------------------------------------------------------------------------------')
        print('Horóscopo na Carreira: \n')
        print(parsed_content.find_all('p')[5].text.strip(), '\n')
        print('-----------------------------------------------------------------------------------------------------')
        print('Palavra do dia: \n')
        print(parsed_content.find_all('p')[6].text.strip(), '\n')
        print('-----------------------------------------------------------------------------------------------------')
        print('Frase do dia: \n')
        print(parsed_content.find_all('p')[7].text.strip(), '\n')
        print('-----------------------------------------------------------------------------------------------------')
        print('Cor do dia: \n')
        print(parsed_content.find_all('p')[8].text.strip(), '\n')
```

```python
#Just input the date on 'DD-MM-YYYY' format (str)
horoscopo_do_dia('09-07-2019')
```
